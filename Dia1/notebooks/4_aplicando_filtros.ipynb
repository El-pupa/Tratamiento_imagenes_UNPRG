```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# LEctura de imagen

```python
path = "/home/stevend/Documents/Notebooks/Dia1/imagenes/lesion.jpg"

img_g=cv2.imread(path,0)

plt.imshow(img_g, cmap="gray")
```

# Binarizacion

```python
_, img_b = cv2.threshold(img_g,200,250, cv2.THRESH_BINARY)

plt.imshow(img_b, cmap="gray")
```

## Aplicamos filtro de suavizado

```python
kernel = np.ones([3,3])/9
```

```python
img_f = cv2.filter2D(img_b,-1,kernel)

plt.imshow(img_f, cmap="gray")
```

## BLUR

```python
img_f2 = cv2.blur(img_b, (10,10))

plt.imshow(img_f2, cmap="gray")
```

## MEdianBLur

### La librera opencv cuenta con una funcin bastante eficaz para corregir este tipo de ruido en imgenes: medianBlur(), en el cual, el kernel recorrer cada pixel de la imagen, e ir asignndole el valor correspondiente a la media de los pixeles que tenga alrededor (comprendidos bajo el kernel). 

```python
img_f3 = cv2.medianBlur(img_b, 9)


plt.imshow(img_f3, cmap="gray")
```

```python
img_f3 = cv2.medianBlur(img_b, 15)
img_f4 = cv2.medianBlur(img_f3, 15)

plt.imshow(img_f4, cmap="gray")
```

## Bordes usando Canny

```python
# Aplicar el filtro Canny
edges = cv2.Canny(img_f4, 200, 250)

plt.imshow(edges, cmap="gray")
```
