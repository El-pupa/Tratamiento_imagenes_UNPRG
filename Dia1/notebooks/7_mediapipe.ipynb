## DEtecci√≥n de la mano

```python
with mp_hands.Hands(
    static_image_mode=True,
    max_num_hands=2,
    min_detection_confidence=0.5) as hands:

    image = cv2.imread('/home/stevend/Documents/Notebooks/Dia1/imagenes/2024-04-22-003618.jpg')
    image = cv2.resize(image, (480,480), interpolation=cv2.INTER_AREA)
    h, w, c = image.shape

    img_f = cv2.flip(image,1)
    img_frgb= cv2.cvtColor(img_f, cv2.COLOR_BGR2RGB)

    results = hands.process(img_frgb)

    print("info mano", results.multi_handedness)

    if not results.multi_hand_landmarks:
        print("No se detectaron manos")
    img_f = cv2.flip(img_frgb,1)

cv2.imshow('MediaPipe Hands', img_f)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Mostrar esqueleto de la mano

```python
with mp_hands.Hands(
    static_image_mode=True,
    max_num_hands=2,
    min_detection_confidence=0.5) as hands:

    image = cv2.imread('/home/stevend/Documents/Notebooks/Dia1/imagenes/2024-04-22-003618.jpg')
    image = cv2.resize(image, (480,480), interpolation=cv2.INTER_AREA)
    h, w, c = image.shape

    img_f = cv2.flip(image,1)
    img_frgb= cv2.cvtColor(img_f, cv2.COLOR_BGR2RGB)

    results = hands.process(img_frgb)

    print("info mano", results.multi_handedness)

    if results.multi_hand_landmarks is not None:    
    # Dibujando los puntos y las conexiones mediante mp_drawing 
        for hand_landmarks in results.multi_hand_landmarks:
            mp_drawing.draw_landmarks(
                img_f, hand_landmarks, mp_hands.HAND_CONNECTIONS,
                mp_drawing.DrawingSpec(color=(255,255,0), thickness=4, circle_radius=5),
                mp_drawing.DrawingSpec(color=(255,0,255), thickness=4))

    if not results.multi_hand_landmarks:
        print("No se detectaron manos")
    img_f = cv2.flip(img_f,1)

cv2.imshow('MediaPipe Hands', img_f)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Mostrar puntos 

```python
with mp_hands.Hands(
    static_image_mode=True,
    max_num_hands=2,
    min_detection_confidence=0.5) as hands:

    image = cv2.imread('/home/stevend/Documents/Notebooks/Dia1/imagenes/2024-04-22-003618.jpg')
    image = cv2.resize(image, (480,480), interpolation=cv2.INTER_AREA)
    h, w, c = image.shape

    img_f = cv2.flip(image,1)
    img_frgb= cv2.cvtColor(img_f, cv2.COLOR_BGR2RGB)

    results = hands.process(img_frgb)

    print("info mano", results.multi_handedness)

    if results.multi_hand_landmarks is not None:    
    # Accediendo al valor de los puntos por su √≠ndice
        index = [4, 8, 12, 16, 20]
        for hand_landmarks in results.multi_hand_landmarks:
            for (i, points) in enumerate(hand_landmarks.landmark):
                if i in index:
                    x = int(points.x * w)
                    y = int(points.y * h)
                    cv2.circle(img_f, (x, y), 3,(255, 0, 255), 3)

    if not results.multi_hand_landmarks:
        print("No se detectaron manos")
    img_f = cv2.flip(img_f,1)

cv2.imshow('MediaPipe Hands', img_f)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
