```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# importing cv2 

# path 
path = r'/home/stevend/Documents/Notebooks/Dia1/imagenes/dedo_gatillo.png'
    
# Reading an image in default mode 
image = cv2.imread(path) 
    
# font 
font = cv2.FONT_HERSHEY_SIMPLEX 
  
# org 
org = (200, 200) 
  
# fontScale 
fontScale = 2
   
# Blue color in BGR 
color = (255, 0, 0) 
  
# Line thickness of 2 px 
thickness = 4
   
# Using cv2.putText() method 
image = cv2.putText(image, 'DEDO', org, font,  
                   fontScale, color, thickness, cv2.LINE_AA) 
   
# Displaying the image 
plt.imshow(image[:,:,::-1])
```
