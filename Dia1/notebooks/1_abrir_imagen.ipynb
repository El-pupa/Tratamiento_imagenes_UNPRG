### Importar librerias

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Abrir imagen usando opencv

```python
%pwd
```

```python
path = "/home/stevend/Documents/Notebooks/Dia1/imagenes/dedo_gatillo.png"

img = cv2.imread(path,cv2.IMREAD_COLOR)
```

## Mostrar imagen en opencv

```python
cv2.imshow("Dedo Gatillo", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Mostrar imagen en matplotlib

```python
plt.imshow(img)
plt.xticks([]), plt.yticks([])
plt.show()
```

## Verificar si la imagen es RGB o BGR

plt.imshow, grafica considerando que la imagen de entrada es RGB, apesar que no sea. La mejor forma de saber si es BGR o RGB es graficandola en su totalidad.

```python
img_aux = img.copy()
img_aux[:,:,0] = 0
img_aux[:,:,1] = 0


plt.imshow(img_aux)
plt.show()
```

```python
plt.imshow(img)
```

## Convertir BGR a RGB

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
```
